# **Rescaling**

```python
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.transform import rescale

img = imread('/content/gdrive/MyDrive/maneesh_passport.jpg')
img_rescaled = rescale(img, scale=(1.1))
plt.subplot(121), imshow(img)
plt.title('Original Image')
plt.subplot(122), imshow(img_rescaled)
plt.title('Rescaled Image')
plt.show()
```

# **Resizing**

```python
from skimage.transform import resize
img1 = img
#resize image
img1_resized = resize(img1, (300, 250))

#plot images
plt.subplot(121), imshow(img1)
plt.title('Original Image')
plt.subplot(122), imshow(img1_resized)
plt.title('Resized Image')
plt.show()
```

# **Flipping**

```python
from numpy import fliplr, flipud
img2 = img
img2_flipud = flipud(img2)
img2_fliplr = fliplr(img2)
plt.subplot(141), imshow(img2)
plt.subplot(142), imshow(img2_fliplr)
plt.subplot(143), imshow(img2)
plt.subplot(144), imshow(img2_flipud)
plt.show()
```
